## 1. MDPs

Discount factor $\gamma$: thigs further in the future will be discounted more, between $[0,1]$, maybe 0.9 or 0.99.

The closer discount factors to zero, the faster things will converge; the closer to one, the longer it imight take for things to converge.

### Example -- leaning robot 

$ MDP (S, A, T, R, \gamma, H)$,

**goal**: $\color{red}maximize$ the discount of accumulated reward:
$$max_\pi E\left[\sum^H_{t=0}\gamma^tR\left(S_t, A_t, S_{t+1}\right)|\pi\right]$$


** state: where everything is in your house as well as where the robot is

** actions: the robot might be able to move somewhere

** transition model: if a robot takes a certain action maybe move forward, what's the chance it actually moves forward, or how far will it move forward 

** reward: get reward if dirt is removed from the floor

** horizen: how many steps we are going to be considering for your robot

## 3. Exact Solution Methods: policy iteration

$\color{red}Q^*(s,a)$: expected utility starting in s, taking action a, and (thereafter) acting optimally.

Not fot a state, but for being in a state and having committed to a specific action in that state.

First term: expected reward on the immediate transition

Second term: expected value at the future state $s'$, this here really is $v^*(s')$, which is the best you can do from state $s'$ so that we need to pick the optimal action $a'$ in that state $s'$.

As you end up with $Q-values$ you can read off the optional action in each state by seeing which one achieves the highest $Q-value$.
